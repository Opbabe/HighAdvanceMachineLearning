Fetch openml [dataset id=37](https://openml.org/search?type=data&sort=runs&status=active&id=37) (diabetes) and build a `SGDClassifier` classifier.

On the test data: show following for your model:

Now retrain the SGDClassifier with different "loss=..." params (see [documentation](https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.SGDClassifier.html) for available options), compute ROC AUC for different options (keep random_state=1237 so that it is reproducible), and find the which one gives the largest value of ROC AUC.


```python
# Get dataset



import pandas as pd
from sklearn.datasets import fetch_openml

diabetes = fetch_openml(data_id=37, as_frame=True)

#this assignment is very complex professor

#i know that not all model can give us probability


# I learned about the limitations of different loss functions in SGDClassifier,
#especially the difference between predict_proba() and decision_function().
```

```python
# Train/Test split
X = diabetes.data
y = diabetes.target

X_train = X.iloc[:576]
X_test = X.iloc[576:]
y_train = y.iloc[:576]
y_test = y.iloc[576:]
```

```python
# Train LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import SGDClassifier

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

#break down the binary
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
y_train_encoded = le.fit_transform(y_train)
y_test_encoded = le.transform(y_test)

model = SGDClassifier(loss='log_loss', random_state=1237)
model.fit(X_train_scaled, y_train_encoded)
```

```python
from sklearn.metrics import confusion_matrix

def print_confusion_matrix(y_true, y_pred):
  return pd.DataFrame(confusion_matrix(y_true, y_pred), columns=['PredNeg', 'PredPos'], index=['TrueNeg', 'TruePos'])
```

```python
# print_confusion_matrix...
y_pred = model.predict(X_test_scaled)
from sklearn.metrics import confusion_matrix

#confusion matrix as a DataFrame
def print_confusion_matrix(y_true, y_pred):
    return pd.DataFrame(confusion_matrix(y_true, y_pred),
                        columns=['Predicted Negative', 'Predicted Positive'],
                        index=['Actual Negative', 'Actual Positive'])

print("Confusion Matrix:")
print(print_confusion_matrix(y_test_encoded, y_pred))




#
```

```python
# Plot precision_recall_curve(y_test)
# Import necessary libraries for plotting
from sklearn.metrics import precision_recall_curve
import matplotlib.pyplot as plt

# checking if it supports predict_proba, otherwise use decision_function
if hasattr(model, "predict_proba"):
    # predictr for the positive class
    y_pred_proba = model.predict_proba(X_test_scaled)[:, 1]
else:
    y_pred_proba = model.decision_function(X_test_scaled)

precision, recall, thresholds = precision_recall_curve(y_test_encoded, y_pred_proba)


#Some model dont use this method so i use deicision function to get the score.



plt.figure()
plt.plot(recall, precision, marker='.', label='Precision-Recall Curve')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend()
plt.show()
```

```python
# Plot roc_curve(y_test)
# Import necessary libraries for ROC curve
from sklearn.metrics import roc_curve, roc_auc_score

#I relize within the positive rate false side by side, i ran into an error.
fpr, tpr, thresholds = roc_curve(y_test_encoded, y_pred_proba)

roc_auc = roc_auc_score(y_test_encoded, y_pred_proba)


plt.figure()
plt.plot(fpr, tpr, marker='.', label=f'ROC Curve (AUC = {roc_auc:.3f})')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend()
plt.show()
```

```python
loss_functions = ['hinge', 'log_loss', 'modified_huber', 'squared_hinge', 'perceptron']

roc_auc_scores = {}




#I learn how to check if the model can do what I want before running the code, got to love how it make me fix
#my mistakes





#simple loop
for loss in loss_functions:
    #current loss function
    model = SGDClassifier(loss=loss, random_state=1237)
    #I need to train the model
    model.fit(X_train_scaled, y_train_encoded)

    if loss == 'log_loss':
        # get probabilities
        y_pred_proba = model.predict_proba(X_test_scaled)[:, 1]
    else:
        # if im using other lose use decision_function
        y_pred_proba = model.decision_function(X_test_scaled)

    roc_auc = roc_auc_score(y_test_encoded, y_pred_proba)
    roc_auc_scores[loss] = roc_auc
    print(f"ROC AUC for loss='{loss}': {roc_auc:.3f}")

best_loss = max(roc_auc_scores, key=roc_auc_scores.get)
print(f"\nBest loss function is '{best_loss}' with ROC AUC of {roc_auc_scores[best_loss]:.3f}")
```
