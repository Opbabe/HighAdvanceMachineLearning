```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, models
import os

# I did not like how the data wasn't loading correctly because the path was wrong.
#I try so many different method but I came to stackoverflow. Zipping issue
#Remember if your training! aninmal folder is nested so double check your work



#I fully understand why I keep running to issue!
#So for residual block this help solve the problem by allowing us for the network to skip some layer
#So at first there was some error about making sure the input and output dimension matching espeically for the stride
#was not 1. So lets say they didnt match, the addition ops would not work because you cant add tensor for different shape



#gemini fix issue, so it may differ from your end, but make sure you add this if you ran into error
#but if you need to fix it just use the gemini auto debug button right under the code text.
data_dir = '/content/animal_data/animal_data'

batch_size = 32

img_height = 224

img_width = 224

# I had to make sure the directory path is correct, I had some trouble with the directory structure.

train_ds = tf.keras.utils.image_dataset_from_directory(
    data_dir,
    validation_split=0.2,
    subset='training',
    seed=123,
    image_size=(img_height, img_width),
    batch_size=batch_size
)

val_ds = tf.keras.utils.image_dataset_from_directory(

    data_dir,


    validation_split=0.2,


    subset='validation',


    seed=123,


    image_size=(img_height, img_width),


    batch_size=batch_size
)



class_names = train_ds.class_names


num_classes = len(class_names)


print("Class names:", class_names)

AUTOTUNE = tf.data.AUTOTUNE
train_ds = train_ds.prefetch(buffer_size=AUTOTUNE)






val_ds = val_ds.prefetch(buffer_size=AUTOTUNE)

# I decided to use some standard data techniques to improve generali
data_augmentation = keras.Sequential([
    layers.RandomFlip('horizontal'),
    layers.RandomRotation(0.1),
    layers.RandomZoom(0.1),
])

# I wrote a custom residual block to use in my models
class ResidualBlock(layers.Layer):
    def __init__(self, filters, strides=1):
        super(ResidualBlock, self).__init__()
        self.conv1 = layers.Conv2D(filters, 3, strides=strides, padding='same', kernel_initializer='he_normal')
        self.bn1 = layers.BatchNormalization()
        self.relu = layers.ReLU()
        self.conv2 = layers.Conv2D(filters, 3, strides=1, padding='same', kernel_initializer='he_normal')


        self.bn2 = layers.BatchNormalization()
        if strides != 1:
            self.shortcut = keras.Sequential([
                layers.Conv2D(filters, 1, strides=strides, padding='same', kernel_initializer='he_normal'),
                layers.BatchNormalization()
            ])
        else:
            self.shortcut = lambda x: x
# 1x1 to adjust when it's needed

    def call(self, inputs):
        x = self.conv1(inputs)
        x = self.bn1(x)
        x = self.relu(x)
        x = self.conv2(x)
# downsample correctly.

        x = self.bn2(x)
        shortcut = self.shortcut(inputs)
        x = layers.add([x, shortcut])
        x = self.relu(x)
        return x
# Lecture notes and Colab

def build_model_from_scratch(input_shape, num_classes):
    inputs = layers.Input(shape=input_shape)
    x = data_augmentation(inputs)
    x = layers.Conv2D(64, 7, strides=2, padding='same', kernel_initializer='he_normal')(x)
    x = layers.BatchNormalization()(x)
    x = layers.ReLU()(x)
    x = layers.MaxPooling2D(pool_size=3, strides=2, padding='same')(x)
    # I had to adjust the filters and strides to ensure the dimensions match up
    x = ResidualBlock(64)(x)
    x = ResidualBlock(128, strides=2)(x)
    #256 becareful
    x = ResidualBlock(256, strides=2)(x)
    x = ResidualBlock(512, strides=2)(x)
    x = layers.GlobalAveragePooling2D()(x)
    outputs = layers.Dense(num_classes, activation='softmax')(x)
    model = keras.Model(inputs, outputs)
    return model



input_shape = (img_height, img_width, 3)

model1 = build_model_from_scratch(input_shape, num_classes)

model1.compile(optimizer='adam',
               loss='sparse_categorical_crossentropy',
               metrics=['accuracy'])

epochs = 10
history1 = model1.fit(
    train_ds,
    validation_data=val_ds,
    epochs=epochs
)

val_loss1, val_acc1 = model1.evaluate(val_ds)
print(f"Validation accuracy for model from scratch: {val_acc1:.4f}")
# Lecture notes and Colab

# I decided to use ResNet50 as the base model
base_model = tf.keras.applications.ResNet50(
    weights='imagenet',
    include_top=False,
    input_shape=(img_height, img_width, 3)
)
base_model.trainable = False
# Lecture notes and Colab

inputs = layers.Input(shape=(img_height, img_width, 3))
x = data_augmentation(inputs)
x = tf.keras.applications.resnet50.preprocess_input(x)
x = base_model(x, training=False)
# I noticed that the output shape of base_model is (None, 7, 7, 2048)
# So, I'll add a residual block with filters=2048
x = ResidualBlock(2048)(x)
x = layers.GlobalAveragePooling2D()(x)
outputs = layers.Dense(num_classes, activation='softmax')(x)
model2 = keras.Model(inputs, outputs)

model2.compile(optimizer='adam',
               loss='sparse_categorical_crossentropy',
               metrics=['accuracy'])

history2 = model2.fit(
    train_ds,

    validation_data=val_ds,

    epochs=epochs
)

val_loss2, val_acc2 = model2.evaluate(val_ds)
print(f"Validation accuracy for pretrained ResNet-50 model: {val_acc2:.4f}")

# Accuracy issue but this is autofill Colab format

print(f"Accuracy of model from scratch: {val_acc1:.4f}")


print(f"Accuracy of pretrained ResNet-50 model: {val_acc2:.4f}")
```
